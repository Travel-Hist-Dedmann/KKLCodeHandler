```python
#converts a KKL part_values_list into a dict.
def KKLcode_to_Dict(input:str)->dict[dict[str]]:
    #put part values in ths dictionary
    operant_dict:dict = {}


    #Version handling
    version_regex:re = re.compile(r"[0-9]{2,3}[*]{2}")
    version:str = re.match(version_regex,input).group(0)
    input:str = input.lstrip(version)
    operant_dict["ver"] = version.rstrip("**")
    print("Version: " + version)
    
    #Imported image handling
    importObjects_regex:re = re.compile(r"/#].+",re.DOTALL)
    importObjects:str = "".join(re.findall(importObjects_regex,input))
    
    if importObjects!= None:
        print(importObjects)
        input:str = input.rstrip("_"+importObjects)
    
  
    count_iter = 0
    count_alpha = 0
    #Split @input by "_" into operant_list
    operant_list:list[str]  = re.split("[_]",input)
   

    alphaNum_regex = re.compile(r"[a-z]{1,2}[0-9]+|[a-z]{2}")
    for i in operant_list:
        count_iter += 1
        alphaNum:str = re.findall(alphaNum_regex,i)[0]
        #if the part value is alphabet + number, change into alphabet + "." + number
        if alphaNum.isalpha()==False:
            
            alpha = re.findall("[a-z]{1,2}",alphaNum)[0]
            num   = re.findall("[0-9]+",alphaNum)[0]
            if len(alpha) == 2:
                dot_added_alphaNumeric = f"{alpha}.{num}"
                i:str = i.replace(alphaNum,dot_added_alphaNumeric)
            else:
                match alpha:
                    case "r" :
                        dot_added_alphaNumeric = f"{alpha}{num}"
                        i:str = i.replace(alphaNum,dot_added_alphaNumeric)
        
        
        #Split element i of operant list by "."    
        i = i.split(".")
        
        part_dict_name = i[0]
        part_vals = i[1:]
        
        #matches:
        
        
    
       
        #Part val_names must be replaced with KKL codenames.
        
        
        part_val_names = [part_dict_name + str(y) for y in range(1,len(i)+1)]
        operant_dict[part_dict_name] = dict(zip(part_val_names,part_vals))  
        
        
    if not(importObjects==""):
        operant_dict["importObjects"] = importObjects 
    print(operant_dict)
    #operant_list:list[list[str]] = [re.split("[.]",x) for x in operant_list] 
    

    output_dict:dict = operant_dict.copy()
 
    print(count_iter,count_alpha)
    return output_dict


"""KKL_part_data["aa"] = [["LeftArm", "_meter"], ["LeftArm2", "_meter"], ["LeftArm2", "_depth"], ["LeftHand", "_menu"], ["LeftHand2", "_meter"], ["RightArm", "_meter"], ["RightArm2", "_meter"], ["RightArm2", "_depth"], ["RightHand", "_menu"], ["RightHand2", "_meter"], ["LeftArmFreeRotation", "_check"], ["RightArmFreeRotation", "_check"]]
KKL_part_data["ia"] = [["Seihuku", "_menu"], ["Seihuku", "_color0"], ["Seihuku", "_color1"], ["Seihuku", "_color2"], ["Seihuku", "_g0"], ["SeihukuLeftArm", "_menu"], ["SeihukuLeftArm", "_color0"], ["SeihukuLeftArm", "_color1"], ["SeihukuRightArm", "_menu"], ["SeihukuRightArm", "_color0"], ["SeihukuRightArm", "_color1"], ["SeihukuBreastOption", "_menu"], ["SeihukuBreastOption", "_color0"], ["SeihukuBreastOption", "_color1"], ["SeihukuBreastOption", "_color2"], ["SeihukuBreastOption", "_reversal2"], ["SeihukuDouOption", "_menu"], ["SeihukuDouOption", "_color0"], ["SeihukuDouOption", "_color1"], ["SeihukuDouOption", "_color2"], ["SeihukuDouOption", "_number"]]
print(len(c["aa"])==len(KKL_part_data["aa"]))
print(len(c["ia"])==len(KKL_part_data["ia"]))
"""
d = KKLcode_to_Dict(KKL_new_data)

d.items()
```
