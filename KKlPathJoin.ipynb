```python
#converts a KKL part_values_list into a path.
def KKLcode_to_Dict(input:str)->dict[list[list[str]]]:
    #Split @input by "_" into operant_list, then Split elements of operant list by "."
    operant_list:list[str]  = (re.split("[_]",input))
    operant_list:list[list[str]] = [(re.split("[.]",x[:])) for x in operant_list] 
    

    output_dict:dict = {}
#iterate over each element of operant list
    for part_values_list in operant_list:
        
        partname_unclean:str =  part_values_list[0]
        
        partname_clean:str = partname_unclean[:2]
        #Put version into it's own category.
        if partname_clean.isnumeric():
            [Version_name,part_name_and_values] = part_values_list[0].split("**")
            output_dict["Version"] = [Version_name][:]
            output_dict[part_name_and_values[:2]] = [part_name_and_values[2:]] + part_values_list[1:] 
            
        #put partnames as keys and part values as value in list.    
        else:
        
            if len(partname_unclean)>0:
                output_dict[partname_clean] =   part_values_list[1:] 
                output_dict[partname_clean].insert(0,partname_unclean[2:])
            else:
                output_dict[partname_clean] = []
        
        #print(f"Version_name:{part_values_list}")
        #print(f"partname:{partname_unclean}")
        #print(f"partname:{partname_clean}")
    print(output_dict)
    #example output: {'Version': '103', 'aa': ['11', '350', '1', '0', '50', '19', '60', '0', '3', '72'], 'ab': ['41', '57', '58', '49'], 'ac':[]...}
   
    return output_dict.copy()
```
